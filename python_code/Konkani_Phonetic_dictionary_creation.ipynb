```python
#code for testing performance of dictionary
def alignTwoFiles(filename):
    dictforkonkaniIPA={}
    
    fileIPA_text=open("../performace_mesearure/data/Phonectic_annaotated_Konkani_IPA_text.txt","r",encoding="utf-8")
    fileKonkani_text=open("../performace_mesearure/data/Phonetic_annaotated_Konkani_text.txt","r",encoding="utf-8")
    filealign=open("../align/"+filename,"w",encoding="utf-8")
    fileIPA_lines=fileIPA_text.readlines()
    fileKonkani_text_lines=fileKonkani_text.readlines()
    
    #print(len(fileKonkani_text_lines))
    #print(len(fileIPA_lines))
    
    for i in range(len(fileKonkani_text_lines)):
        
        
        konkani_text_words=fileKonkani_text_lines[i].rstrip(" \n").split()
        fileIPA_words=fileIPA_lines[i].rstrip(" \n").split()
        #'''
        
        if len(konkani_text_words)!=len(fileIPA_words):
            print("line number", i)
            print(len(konkani_text_words))
            print(konkani_text_words)
            print(len(fileIPA_words))
            print(fileIPA_words)
        #'''
        
        for ind in range(len(konkani_text_words)):
            concat=konkani_text_words[ind]+"#"+fileIPA_words[ind]
            if concat in dictforkonkaniIPA:
                dictforkonkaniIPA[concat]=dictforkonkaniIPA.get(concat)+1
            else:
                dictforkonkaniIPA[concat]=1
        
    for key in dictforkonkaniIPA:
        filealign.write(key)
        #filealign.write("#"+value)
        filealign.write("\n")
    filealign.close() 
    fileIPA_text.close()
    fileKonkani_text.close()
    
alignTwoFiles("alignkonkaniText_IPA.txt")
```

```python
def searchforIPAtranscription():
    fileIPATransoutput=open("../performace_mesearure/data/phonetic_dictionary.txt","r",encoding="utf8")
    fileGS_IPAtext=open("../performace_mesearure/data/GS_IPAtext.txt","w",encoding="utf8")
    fileGS_IPAtext_NOTexisting=open("../performace_mesearure/data/GS_IPAtext_NOtexisting.txt","w",encoding="utf8")
    filealigned=open("../align/alignkonkaniText_IPA.txt","r",encoding="utf8")
    fileIPATransoutput_lines=fileIPATransoutput.readlines()
    filealigned_lines=filealigned.readlines()
    
    list_konk_text=[]
    list_IPA_text=[]
    for line in filealigned_lines:
        line=line.rstrip("\n")
        word=line.split("#")
        list_konk_text.append(word[0])
        list_IPA_text.append(word[1])
        
    for line in fileIPATransoutput_lines:
        word=line.split(" ")
        if word[0] in list_konk_text:
            ActualIPAword=list_IPA_text[list_konk_text.index(word[0])]
            fileGS_IPAtext.write(word[0]+" ")
            fileGS_IPAtext.write(ActualIPAword+" ")
            fileGS_IPAtext.write(word[1])
        else:
            fileGS_IPAtext_NOTexisting.write(word[0]+"\n")
    
    fileGS_IPAtext.close()   
    fileGS_IPAtext_NOTexisting.close()
    filealigned.close()
    fileIPATransoutput.close()

searchforIPAtranscription()
```

```python
def accuracy():
    count=0
    fileGS_IPAtext=open("../performace_mesearure/data/GS_IPAtext.txt","r",encoding="utf8")
    fileGS_err=open("../performace_mesearure/data/err.txt","w",encoding="utf8")
    fileGS_IPAtext_lines=fileGS_IPAtext.readlines()
    for line in fileGS_IPAtext_lines:
        line=line.rstrip("\n")
        word=line.split(" ")
        if word[1]==word[2]:
            count=count+1
        else:
            fileGS_err.write(line+"\n")
    fileGS_err.close()
    print("count",count)
    print("len(fileGS_IPAtext_lines)",len(fileGS_IPAtext_lines))
    acc=count/len(fileGS_IPAtext_lines)*100
    print("Word Acc",acc)
    
accuracy()
            
```
